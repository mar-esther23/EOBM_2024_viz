## Conjuntos de datos ordenados

Los principios de los **datos ordenados** (tidy data) proporcionan una forma estándar de organizar los valores de datos dentro de un conjunto de datos. Un estándar hace que la limpieza inicial de datos sea más fácil, ya que no es necesario empezar desde cero y reinventar la rueda cada vez. El estándar de datos ordenados ha sido diseñado para facilitar la exploración inicial y el análisis de los datos, y para simplificar el desarrollo de herramientas de análisis de datos que funcionen bien juntas. Estos principios  fueron dados por Hadley Wickham ([artículo original](http://vita.had.co.nz/papers/tidy-data.pdf)).

Una **variable** es una medida o un atributo que puede ser asignado a un objeto o individuo. Por ejemplo, la edad, el peso, la altura, el género, la fecha de nacimiento, etc. son variables que pueden ser asignadas a una persona. Una variable puede ser cuantitativa (en la que los valores pueden ser medidos en una escala numérica) o cualitativa (en la que los valores son categóricos, es decir, no pueden ser medidos en una escala numérica). En ambos casos, la variable se utiliza para describir algo y para analizar datos.

Un **valor** es una medida o atributo específico asociado a una variable. Por ejemplo, si una variable es la fecha de nacimiento, un valor podría ser 10 de mayo. Si la variable es el peso, un valor podría ser 80 kg. Los valores son los valores concretos o específicos asociados con una variable en un momento dado.

Una **observación** es una serie de medida o registro de una variable o atributo en un momento o en una unidad específica. Por ejemplo, la observación de la altura y peso de una persona en un momento específico. En un estudio o experimento, las observaciones se recopilan y se utilizan para extraer información y conclusión sobre la relación entre variables o para describir una población o un fenómeno.

Un **dato** es una unidad de información, ya sea numérica o categórica, que se recolecta para describir o medir un fenómeno o una característica de un objeto o sujeto. Por ejemplo, la edad, el género, la estatura, el peso, la fecha de nacimiento, entre otros, son datos que se pueden recopilar sobre una persona. Los datos se utilizan para analizar patrones y tendencias, tomar decisiones basadas en información y generar conocimientos nuevos a partir de la información existente.




## Tipos de datos

Existen varios tipos de datos:

* Númericos: valores contables discretos o continuos
* Binarios: verdadero o falso
* Categoricos: número limitado de categorias
* Texto
* Otros: Fecha, Coordenadas, etc

Los datos no siempre están completos, es decir, a veces desconocemos el valor de un dato. Esto se puede representar de varias formas. En el caso de pandas se usa NaN para representar los datos faltantes o desconocidos.

Al cargar un conjunto de datos es importante asegurarse que cada columna (variable) este en el formato adecuado.

```python
from pandas.api.types import CategoricalDtype

df = pd.read_csv(in_dir+'data_raw/defuncion_nacion.csv', dtype=str)

# Año
df['Año'] = df['Año'].apply(lambda x: int(x))
# Edad
cat_edad = list( df['Edad'].unique() )
cat_edad = CategoricalDtype(categories=cat_edad, ordered=True)
df['Edad'] = df['Edad'].astype(cat_edad)
# Diagnostico: desgloce respiratorias
diag_resp = ['Influenza (gripe)', 'Neumonía',
       'Otras infecciones agudas de las vías respiratorias inferiores',
       'Enfermedades crónicas de las vías respiratorias inferiores',
       'Resto de enfermedades del sistema respiratorio']
df_resp = df[df['Diagnostico'].isin(diag_resp)]
df = df[~df['Diagnostico'].isin(diag_resp)]
# Sexo
df['Sexo'] = df['Sexo'].astype('category')
# Frecuencia
df['Frecuencia'] = df['Frecuencia'].apply(lambda x: int(x))

df.dtypes
```

## Profile de datos

Un **profiler de datos** en Python es una herramienta o librería que se utiliza para analizar y generar un informe detallado sobre las características de un conjunto de datos. Su propósito principal es ayudar a los usuarios a comprender mejor los datos con los que están trabajando. Esto incluye información sobre la estructura, calidad y distribución de los datos, como:

- **Resumen estadístico** de las columnas: media, mediana, moda, desviación estándar, etc.
- **Valores faltantes**: cuántos valores faltan en cada columna.
- **Distribución de datos**: cómo se distribuyen los valores en las columnas.
- **Valores atípicos (outliers)**: si hay valores que se desvían significativamente del resto.
- **Tipos de datos**: qué tipos de datos (numéricos, categóricos, fechas) están presentes.
- **Correlaciones**: análisis de correlación entre variables.

Estas herramientas son muy útiles para realizar un **análisis exploratorio de datos (EDA)** y comprender el estado inicial de un dataset antes de realizar tareas más avanzadas, como la limpieza de datos o el modelado.

## Codificación visual

Se representa la información de manera gráfica mediante elementos visuales

Datos → Elementos visuales

Elementos visuales: colores, formas, tamaños y posiciones, etc.

La codificación depende de los datos, sus tipos y la jerarquía interna.
Generar codificación:
* Determina variables y tipo de dato
* Jerarquiza variables
* Asigna mapeo

**Elementos visuales posicionales**
* Puntos:
    * dos escalas cuantitativas
    * mediciones independientes
* Líneas:
    * una de las escalas es secuencial, por ejemplo tiempo
    * resaltar forma o cambios
    * líneas suaves y continuas indican proceso continuo
    * usa puntos y líneas para resaltar mediciones
    * NO usar para categóricos
* GráficaGráficaBarras:
    * Representar y comparar valores en categorías
    * Series de tiempo si interesa comparación entre puntos
    * Siempre empieza en cero (si eje no empieza en cero usar puntos)

**Elementos visuales retinales:**
* Tamaño, orientación, tono, saturación, forma y textura

## Gráfica: Defunciones por diagnóstico de 2010 a 2024

```python
# Obtener tabla pivote
df_ad = df.pivot_table(index='Año', columns='Diagnostico',
                       values='Frecuencia', aggfunc=sum, fill_value=0)
# Acortar nombres de diagnósticos
df_ad = df_ad.rename(diag_corto, axis=1)
# Ordenar de mayor a menor
df_ad.loc['total']= df_ad.sum()
df_ad = df_ad.sort_values(by='total', axis=1, ascending=False)
df_ad = df_ad.drop('total')
# Sumar baja frecuencia
n_small = 8
otras = df_ad.iloc[:,n_small-1:].sum(axis=1)
df_ad = df_ad.iloc[:,0:n_small-1]
df_ad.insert(0, 'Otras', otras)
df_ad
# graficar
dic_color = {'Otras':'#8dd3c7', 'Circulatorio':'#ffffb3', 'Endocrinas y metabólicas':'#bebada',
             'Tumores':'#fccde5', 'Externas':'#80b1d3', 'Digestivo':'#fdb462',
             'Respiratorio':'#b3de69', 'Emergente':'#fb8072'}
df_ad.plot.area(
            color=dic_color,
            xlabel='Año',
            ylabel='Defunciones (millones)',
            title='Defunciones por diagnóstico de 2010 a 2024'
            )
plt.show()
```
